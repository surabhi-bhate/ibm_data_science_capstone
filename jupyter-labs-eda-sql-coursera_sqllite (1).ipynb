```python
import sqlite3
import pandas as pd

# Connect to SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# Question 1: Retrieve up to 20 records from the table
query1 = "SELECT * FROM SPACEXTABLE LIMIT 20;"
df1 = pd.read_sql_query(query1, con)
print("Question 1 - First 5 rows:")
print(df1.head())

# Question 2: Display the minimum payload mass
query2 = "SELECT MIN(payload_mass__kg_) as MinPayloadMass FROM SPACEXTABLE;"
df2 = pd.read_sql_query(query2, con)
print("\nQuestion 2 - Minimum Payload Mass:")
print(df2)

# Question 3: Total payload mass with alias Total_Payload_Mass
query3 = "SELECT SUM(payload_mass__kg_) AS Total_Payload_Mass FROM SPACEXTABLE;"
df3 = pd.read_sql_query(query3, con)
print("\nQuestion 3 - Total Payload Mass:")
print(df3)

# Question 4: Mission outcome counts for each launch site
query4 = """
SELECT launch_site, COUNT(mission_outcome) AS MISSION_OUTCOME_COUNT
FROM SPACEXTABLE
GROUP BY launch_site;
"""
df4 = pd.read_sql_query(query4, con)
print("\nQuestion 4 - Mission Outcome Counts by Launch Site:")
print(df4)

# Question 5: Unique launch sites in the table
query5 = "SELECT DISTINCT launch_site FROM SPACEXTABLE;"
df5 = pd.read_sql_query(query5, con)
print("\nQuestion 5 - Unique Launch Sites:")
print(df5)

con.close()
```

```python
# ----------------------------
# Import Libraries
# ----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------
# Load Dataset
# ----------------------------
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv"
df = pd.read_csv(url)

# Inspect columns
print(df.columns)
# Expected columns:
# ['Date', 'Time (UTC)', 'Booster_Version', 'Launch_Site', 'Payload', 
#  'PAYLOAD_MASS__KG_', 'Orbit', 'Customer', 'Mission_Outcome', 'Landing_Outcome']

# ----------------------------
# 1. Count of Launches per Site (Bar Plot)
# ----------------------------
plt.figure(figsize=(10,5))
launch_counts = df['Launch_Site'].value_counts()
sns.barplot(
    x=launch_counts.index,
    y=launch_counts.values,
    hue=launch_counts.index,
    palette=sns.color_palette("Set2", len(launch_counts)),
    dodge=False,
    legend=False
)
plt.title("Number of Launches per Launch Site")
plt.xlabel("Launch Site")
plt.ylabel("Number of Launches")
plt.xticks(rotation=45)
plt.show()

# ----------------------------
# 2. Flight Number vs Payload Mass (Scatter Plot, colored by Mission Outcome)
# ----------------------------
# Fix palette dynamically
unique_outcomes = df['Mission_Outcome'].unique()
palette = sns.color_palette("Set2", len(unique_outcomes))

plt.figure(figsize=(12,5))
sns.scatterplot(
    data=df,
    x='Date',  # Flight Number not available, use Date for x-axis
    y='PAYLOAD_MASS__KG_',
    hue='Mission_Outcome',
    palette=palette
)
plt.title("Payload Mass vs Launch Date (Mission Outcome Color-coded)")
plt.xlabel("Launch Date")
plt.ylabel("Payload Mass (kg)")
plt.xticks(rotation=45)
plt.legend(title="Mission Outcome")
plt.show()

# ----------------------------
# 3. Flight Number vs Launch Site (Scatter Plot)
# ----------------------------
plt.figure(figsize=(12,5))
sns.scatterplot(
    data=df,
    x='Date',  # Flight Number replaced by Date
    y='Launch_Site',
    hue='Mission_Outcome',
    palette=palette
)
plt.title("Launch Site vs Launch Date (Mission Outcome Color-coded)")
plt.xlabel("Launch Date")
plt.ylabel("Launch Site")
plt.xticks(rotation=45)
plt.legend(title="Mission Outcome")
plt.show()

# ----------------------------
# 4. Payload Mass Distribution (Boxplot)
# ----------------------------
plt.figure(figsize=(10,5))
sns.boxplot(
    x='Launch_Site',
    y='PAYLOAD_MASS__KG_',
    data=df,
    palette="Set3"
)
plt.title("Payload Mass Distribution per Launch Site")
plt.xlabel("Launch Site")
plt.ylabel("Payload Mass (kg)")
plt.xticks(rotation=45)
plt.show()

# ----------------------------
# 5. Success/Failure Pie Chart (per Launch Site)
# ----------------------------
for site in df['Launch_Site'].unique():
    site_data = df[df['Launch_Site'] == site]
    counts = site_data['Mission_Outcome'].value_counts()
    plt.figure(figsize=(6,6))
    plt.pie(
        counts,
        labels=counts.index,
        autopct='%1.1f%%',
        colors=sns.color_palette("Set2", len(counts))
    )
    plt.title(f"Mission Outcome Distribution for {site}")
    plt.show()

# ----------------------------
# 6. Payload vs Landing Outcome (Scatter Plot)
# ----------------------------
unique_landing = df['Landing_Outcome'].unique()
palette_landing = sns.color_palette("Set1", len(unique_landing))

plt.figure(figsize=(12,5))
sns.scatterplot(
    data=df,
    x='PAYLOAD_MASS__KG_',
    y='Launch_Site',
    hue='Landing_Outcome',
    palette=palette_landing
)
plt.title("Payload Mass vs Launch Site (Landing Outcome Color-coded)")
plt.xlabel("Payload Mass (kg)")
plt.ylabel("Launch Site")
plt.legend(title="Landing Outcome")
plt.show()
```

```python
# Calculate success rate for each orbit type
success_rate = df.groupby("Orbit")["Mission_Outcome"].apply(
    lambda x: (x.str.contains("Success").sum() / len(x)) * 100
).reset_index(name="Success_Rate")

# Plot bar chart
plt.figure(figsize=(12,6))
sns.barplot(data=success_rate, x="Orbit", y="Success_Rate", palette="viridis")

plt.title("Success Rate by Orbit Type")
plt.xlabel("Orbit Type")
plt.ylabel("Success Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract year
df['Year'] = df['Date'].dt.year

# Define success as 1 and failure as 0
df['Success'] = df['Mission_Outcome'].apply(lambda x: 1 if 'Success' in x else 0)

# Group by year and calculate average success rate
yearly_success = df.groupby('Year')['Success'].mean().reset_index()

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(yearly_success['Year'], yearly_success['Success'], marker='o', linestyle='-', color='blue')

plt.title("Yearly Average Success Rate of SpaceX Missions")
plt.xlabel("Year")
plt.ylabel("Average Success Rate")
plt.grid(True)
plt.show()
```
